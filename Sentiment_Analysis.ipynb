```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv(r"IMDB-Dataset.csv")
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df['sentiment'].value_counts())
```

```python
df = df.drop_duplicates()
```

```python
print("Dataset after removing duplicates: ", df.shape)
```

```python
X = df['review']
Y = df['sentiment']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 42, stratify = Y)
```

```python
print("Training Examples: ", len(X_train))
print("Testing Examples: ", len(X_test))
```

```python
model = Pipeline([              # Reduces the lines of code we need to write and errors by handling the middle steps on its own.

    (
        "tfidf",                # "tfidf" and "classifier" are just the names we give the different steps in the pipeline and
        TfidfVectorizer(        # can be anything. They are useful when we want to alter a value/setting in a step later on.
            max_features = 15000
        )
    ),
    (
        "classifier",
        LogisticRegression(
            max_iter = 300,
            solver = "liblinear"    # Solver is the method used for calculation of the result.
        )
    )

])
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, Y_pred)

print("Accuracy: ", accuracy)
print("Classification report: ")
print(classification_report(Y_test, Y_pred))
print("Confusion Matrix: ")
print(confusion_matrix(Y_test, Y_pred))
```

```python
new_review = ["This movie was awesome."]

new_sentiment = model.predict(new_review)
print("Sentiment: ", new_sentiment[0])  # Using [0], prints the only result. Otherwise, it prints an array like ['positive'].
```

```python
new_reviews = [
    "The performances were excellent and I enjoyed every minute.",
    "The movie was dull, predictable, and far too long.",
    "It was not perfect, but it was still quite entertaining."
]

predictions = model.predict(new_reviews)

# zip packages corresponding new_reviews and predictions together. So first review is packaged with first sentiment and so on.

for review, sentiment in zip(new_reviews, predictions):
    print("Review:", review)
    print("Prediction:", sentiment)
    print()
```
